# Data Array Components
Values (values): Actual valuse stored in Numpy
Dimensions (dims): Named axes e.g time, latitude and longitued
Coordinates (coords) : Labels for values e.g specific tiem or geographic location 
Attributes (attrs) : Meta associated with the data.

# Indexing and Selecting Data

```python
seleted_data=temperature.sel(time='2013-01-01',lat=40.0, lon=260.0)
seleted_data
```

```python
# Select data across a range of times
time_slice = temperature.sel(time=slice("2013-01-01","2013-01-31"))
time_slice
```

# Performing Operations on DataArray

```python
mean_temperature=temperature.mean(dim='time')
mean_temperature
```

```python
# subtract the mean temperature from the original data
anomalies = temperature - mean_temperature
anomalies
```

# Visualizatio with Xarray

```python
mean_temperature.plot(cmap = 'jet',figsize=(10,6))
plt.title("Main temperatue")
plt.xlabel("Longitude")
plt.ylabel("Latitude")
```

```python
# Plot a time series for a specific location
temperature.sel(lat=40.0,lon=260.0).plot()
plt.show()
```

# Advance Indexing : Label Vs Position-Based Indexing

```python
# Select all data from May 2013
ds.sel(time ="2013-05")
```

```python
# Select the first time step, second latitude, and third longitude
ds.air.isel(time = 0, lat=2, lon=3)
```

# High Level Computation with Xarray

```python
season_mean = ds.groupby('time.season').mean()
season_mean.air.plot(col='season')
```

# Computation with Weights

```python
cell_area=xr.ones_like(ds.air.lon) # place holder for actual area calculation
weighted_mean=ds.weighted(cell_area).mean(dim=['lon','lat'])
weighted_mean.air.plot()
```

# Rolling Window Opeation

```python
ds.air.isel(time=0).rolling(lat=5,lon=5).mean().plot()
```

```python
plt.figure(figsize=(10,6))
temperature =ds['air'].sel(lat=40,lon=260)
temperature.plot(label  =   'Original')
#Apply rolling mean smooting with a windows size of 20
smoothed_temperature=temperature.rolling(time=20,  center = True).mean()

#plot the smoothed data
smoothed_temperature.plot(label ='Smoothed')
 #Add a title and labels
plt.title("Temperature Time Series (lat=40.0, lon=260.0)")
plt.xlabel("Time")
plt.ylabel("Temperature (K)")

# Add a legend
plt.legend()

# Show the plot
plt.show()
```
